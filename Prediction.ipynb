```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Dense, Embedding, SimpleRNN
```

```python
word_index=imdb.get_word_index()
##word_index

reversed_word_index={vale:key for key,vale in word_index.items()}
reversed_word_index
```

```python
model = load_model('SimpleRNNIMDB.h5')
model.summary()
```

```python
model.get_weights()
```

```python
## step2 helper finction
## Function to decode the review
maxlen = 500

def  decode_review(enocded_review):
    return ' '.join([reversed_word_index.get(i-3,'?') for i in enocded_review])
def pre_processing_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2)+3 for word in words]
    padded_review=sequence.pad_sequences([encoded_review], maxlen=maxlen)
    return padded_review

```

```python
### Prediction function

def predict_sentiment(reviwe):
    preprocessingtext=pre_processing_text(reviwe)
    prediction=model.predict(preprocessingtext)

    sentiment='Positive' if prediction[0]>0.5 else 'Negative'
    return sentiment,prediction[0][0]

```

```python
example_reviews = [
    "I loved the movie! It was so exciting to watch!",
    "I hated the movie. It was boring and dull.",
    "I didn't like the movie at all. It was terrible."
]

for review in example_reviews:
    sentiment, score = predict_sentiment(review)
    print(f"Review: {review}\nSentiment: {sentiment}, Score: {score:.2f}\n")
    
```
